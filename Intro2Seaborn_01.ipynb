```python
height = np.array([62, 64, 69, 75, 66, 68, 65, 71, 76, 73])
weight = np.array([120, 136, 148, 175, 137,165, 154, 172, 200, 187])

sns.scatterplot(x=height, y=weight)
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x="total_bill",
                y="tip",
                data=tips,
                hue="smoker")
plt.show()
```

```python
sns.scatterplot(x="total_bill",
                y="tip",
                data=tips,
                hue="smoker",
                hue_order=["Yes", "No"])
plt.show()
```

```python
hue_colors = {"Yes": "blue",
              "No": "orange"}

sns.scatterplot(x="total_bill",
                y="tip",
                data=tips,
                hue="smoker",
                hue_order=["Yes", "No"],
                palette=hue_colors)
plt.show()
```

```python
sns.countplot(x="smoker",
              data=tips)
plt.show()
```

```python
sns.countplot(x="smoker",
              data=tips,
              hue='sex')
plt.show()
```
