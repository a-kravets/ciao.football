```python
# Load csv

file = r'D:\Python\Code\ciao.football\Serie A\mc.csv'
df = pd.read_csv(file)
df
```

```python
pd.set_option("display.max.columns", None)

df_pair = df[(df['#'] == 2) | (df['#'] == 11)].reset_index()

'''
df_walker = df.loc[df['#'] == 2].reset_index()
df_walker

df_zinchenko = df.loc[df['#'] == 11].reset_index()
df_zinchenko
'''

df_pair
```

```python
#df_zinchenko = df_zinchenko[['xG', 'npxG', 'xA', 'SCA', 'Cmp', 'Att', 'Cmp%', 'Carries']]
#df_walker = df_walker[['xG', 'npxG', 'xA', 'SCA', 'Cmp', 'Att', 'Cmp%', 'Carries']]

df_pair = df_pair[['Player', 'Touches', 'Press', 'Tkl', 'Int', 'Blocks', 'xA', 'SCA', 'Cmp', 'Att', 'Cmp%', 'Carries', 'Prog.1', 'Succ', 'Att.1']]
```

```python
df_pair
```

```python
#get parameters
params = list(df_pair.columns)
params = params[1:]
params
```

```python
#add ranges to list of tuple pairs
ranges = []
a_values = []
b_values = []

for x in params:
    a = min(df_pair[params][x])
    a = a - (a*.25)
    
    b = max(df_pair[params][x])
    b = b + (b*.25)
    
    ranges.append((a,b))
```

```python
for x in range(len(df_pair['Player'])):
    if df_pair['Player'][x] == 'Oleksandr Zinchenko\Oleksandr-Zinchenko':
        a_values = df_pair.iloc[x].values.tolist()
    if df_pair['Player'][x] == 'Kyle Walker\Kyle-Walker':
        b_values = df_pair.iloc[x].values.tolist()

a_values = a_values[1:]
b_values = b_values[1:]
        
values = [a_values,b_values]
```

```python
title = dict(
    title_name='Oleksandr Zinchenko',
    title_color = 'red',
    subtitle_name = 'Man City',
    subtitle_color = 'red',
    title_name_2='Kyle Walker',
    title_color_2 = 'blue',
    subtitle_name_2 = 'Man City',
    subtitle_color_2 = 'blue',
    title_fontsize = 18,
    subtitle_fontsize=15
)

endnote = '@ciao.football\ndata via FBref / Statsbomb'
```

```python
radar = Radar()

fig,ax = radar.plot_radar(ranges=ranges,params=params,values=values,
                         radar_color=['red','blue'], title=title, endnote=endnote,
                         alphas=[.75,.6], compare=True)
```
